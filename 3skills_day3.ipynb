**Functions of list**

append(x) – adds one item to the end

extend(iterable) – adds multiple items

insert(i, x) – inserts item at a specific index

remove(x) – removes first matching value

pop(i) – removes and returns item at index (last by default)

clear() – removes all items

index(x) – returns index of first match

count(x) – counts occurrences of a value

sort() – sorts the list in place

reverse() – reverses the list

**Tuple methods**
count(x) – number of times a value appears

index(x) – position of the first occurrence

Example

**list**

```python
list=[10,20,30]
print("original list:",list)

#append used for single element in list
list.append(40)
print("after append:",list)

#extend used to add multiple elments in list
list.extend([50,60])
print("after extend :",list)

#insert
list.insert(2,25)
print("after insert:",list)

#remove- remove elemnet by giving value
list.remove(20)
print("after remove:",list)

#pop -remove by giving index val
list.pop(3)
print("after pop:",list)

#sort
list.sort()
print("after sort:",list)

#reverse
list.reverse()
print("after reverse:",list)
```

**Tuple function**

```python
t=(10,20,30,40,50,60)
print("original tuple",t)
#max
print("maximum val:",max(t))
#min
print("minimun val:",min(t))
#sum
print("sum of elemnets:",sum(t))
#membership
print("is 70 in tuple?",70 in t)
```
