```python
# create a dict
student={"name":"aarushi","age":"18","course":"Btech","marks":"90"}
print("orginal dictionary:")
print(student)

#Accessing values
print("\naccessing values:")
print("name:",student["name"])

#updating marks
print("\n updating marks:")
student["marks"]=95
print(student)

#adding college
print("\n adding college:")
student["college"]="xyz"
print(student)

#keys
print("\n keys")
print(student.keys())

#values
print("\n values")
print(student.values())

#items
print("\n items")
print(student.items())

#get
print("\n get")
print(student.get("name"))

#default
student.setdefault("city","lucknow")
print("\n after setdefault():")
print(student)

#pop
student.pop("age")
print("\n after pop")
print(student)
#pop items
student.popitem()
print("\n after popitem()")
print(student)

#delete
del student["course"]
print("\n after del course")
print(student)
```
