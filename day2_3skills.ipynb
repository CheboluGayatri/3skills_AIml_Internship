**3.Operators**
operators are symbol that perform operations that are addition,sub,mul,div

```python
#1.Arthematic operator
a=10
b=5
print("addition:",a+b)
print("substraction:",a-b)
print("Multiplication:",a*b)
print("Division:",a/b)
print("Modules:",a%b)
print("floordivision:",a//b)
```

```python
#2.Comparision Operator
a=10
b=5
print("a is greater:",a>b)
print("a is less:",a<b)
```
